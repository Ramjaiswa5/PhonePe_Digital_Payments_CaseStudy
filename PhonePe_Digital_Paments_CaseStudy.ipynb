```python
def missing_column(df):
    # Count missing values
    missing_count = df.isnull().sum()
    # Calculate percentage
    missing_percent = (missing_count / len(df)) * 100
    # Pick only columns with missing values
    missing_percent = missing_percent[missing_percent > 0]

    if not missing_percent.empty:
        print(missing_percent.map("{:.2f}".format))
```

```python
# 3. Highlight which column has the highest percentage of missing values in each dataset.

def highest_missing_column(df, dataset_name="Dataset"):
    # Count missing values
    missing_count = df.isnull().sum()
    # Calculate percentage
    missing_percent = (missing_count / len(df)) * 100
    # Pick only columns with missing values
    missing_percent = missing_percent[missing_percent > 0]

    if not missing_percent.empty:
        max_col = missing_percent.idxmax()
        max_val = missing_percent.max()
        print(f"{dataset_name}: '{max_col}' has the highest missing values ({max_val:.2f}%)")
    else:
        print(f"{dataset_name}: No missing values found.")
```

```python
# Group by state and sum up relevant columns
district_summary = (District_Txn_Users.groupby("State")[["Transactions", "Amount (INR)", "Registered Users"]].sum().reset_index())
display(district_summary)
```

```python
# 2. Compare the results with the corresponding values at the state level to ensure they match.

state_summary = (State_Txn_Users.groupby("State")[["Transactions", "Amount (INR)", "Registered Users"]].sum().reset_index())
display(state_summary)
```

```python
comparison = pd.merge(district_summary, state_summary, on="State", suffixes=("_from_districts", "_from_states"))

# Calculate differences
comparison["Transactions_diff"] = comparison["Transactions_from_districts"] - comparison["Transactions_from_states"]
comparison["Amount_diff"] = comparison["Amount (INR)_from_districts"] - comparison["Amount (INR)_from_states"]
comparison["Users_diff"] = comparison["Registered Users_from_districts"] - comparison["Registered Users_from_states"]

display(comparison[["State","Transactions_diff","Amount_diff","Users_diff"]])
```

```python
# 3. Display any discrepancies found between the district-level and state-level data.

discrepancies = comparison[
    (comparison["Transactions_from_districts"] != comparison["Transactions_from_states"]) |
    (comparison["Amount (INR)_from_districts"] != comparison["Amount (INR)_from_states"]) |
    (comparison["Registered Users_from_districts"] != comparison["Registered Users_from_states"])
]
# Displaying only relevant columns
display(discrepancies[["State", "Amount (INR)_from_districts", "Amount (INR)_from_states"]])
```
